```python
import cv2
import numpy as np

vidlive = cv2.VideoCapture(0)
vidlive.set(3,640)
vidlive.set(4,400)
vidlive.set(10,160)

myColors = [[133,56,0,159,156,255],[57,76,0,100,255,255],[0,179,0,255,0,255]]  #  Blue and green,yellow
myColorValues = [[51,51,255],          ## BGR (not RGB) and we can get values from https://www.rapidtables.com/web/color/RGB_Color.html
                 [255,153,51],
                 [255,255,0]]   # add this variable as input to my function ( order should be as myColors ), then create counter

myPoints =  []  ## [x , y , colorId ] we will loop this list to check x,y and draw color at this time

def get_color(img,myColors,myColorValues):
    imgHSV= cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    count=0
    newPoints= []
    for color in myColors:  # Loop for color 
        lower= np.array(color[0:3])
        upper=np.array(color[3:6])
        mask = cv2.inRange(imgHSV,lower, upper)  # filter bet lower and upper so we need to detect right values by visual them
#         cv2.imshow(str(color[0]),mask)
        x,y=  getContours(mask)  # retun new_x and y (top of contour)
#         cv2.circle(newimage,(x,y),15,(0,0,255),cv2.FILLED)  # red color as test ( center of bounding box, not center of contour)
        cv2.circle(newimage,(x,y),15,myColorValues[count],cv2.FILLED)  #, 15 is radius
        if x!=0 and y!=0:  # to avoid appending zeros in case of area less 500 
            newPoints.append([x,y,count])
        count +=1 
    return newPoints  # every time send list of these points to below 
        

def getContours(img):   # Function to take image (mask) and draw contour for different shapes
    contours,hierarchy = cv2.findContours(img,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)  # Find external, no all Contours
    x,y,w,h = 0,0,0,0   # in case of less than 500
    for cnt in contours:
        area = cv2.contourArea(cnt)  # find area: The function computes a contour area. Similarly to moments 
        if area>500:   # to avois small area and noise 
            cv2.drawContours(newimage, cnt, -1, (255, 0, 0), 3)  # -1 all Contours, if 1 means index =1
            peri = cv2.arcLength(cnt,True)   # True means curve is closed : Calculates a contour perimeter or a curve length.
            approx = cv2.approxPolyDP(cnt,0.02*peri,True)  #It approximates a contour shape to another shape with less number of vertices depending upon the precision we specify
            x, y, w, h = cv2.boundingRect(approx)  
    return x+w//2,y  # we need to send top, not center 

def drawOnCanvas(myPoints,myColorValues):
    for point in myPoints:
        cv2.circle(newimage, (point[0], point[1]), 10, myColorValues[point[2]], cv2.FILLED)
        # we need to take this color from function get_color and send them to this function
        
while True:
    ret, frame = vidlive.read() 
    newimage = frame.copy()  # new image to have all results
    if ret == True: 
        newPoints = get_color(frame,myColors,myColorValues) # get the color to send to drawOnCanvas
        if len(newPoints)!=0:  
            for newP in newPoints:  # here to convert list of new points to points and then add them to orginal list of myPoints
                myPoints.append(newP) # we can't add list to list, so we convert newPoints firstly to points
        if len(myPoints)!=0:  # here to draw all points if not zero 
            drawOnCanvas(myPoints,myColorValues)
        
        cv2.imshow("Output Video",newimage)      
        if cv2.waitKey(1) & 0xFF == ord('q'):
                break
    else:
        break 
```
